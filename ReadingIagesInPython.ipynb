```python
#scikit-image
from skimage import io, img_as_float, img_as_ubyte
```

```python
image = io.imread("Untitled design (11).png")
```

```python
print(type(image))
image_float = img_as_float(image)
print(image_float)

# As you can see from the output, 'image' is already a numpy.ndarray.
# The skimage function 'img_as_float' processes this numpy array,
# converting its data type to float, confirming that scikit-image
# handles images as arrays.
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
plt.show()
```
